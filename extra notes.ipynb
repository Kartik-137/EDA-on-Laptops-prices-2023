`SSD Vs HDD` Ref: **Google and IBM article**
- An HDD is a “non-volatile” storage drive, which means it can retain the stored data even when no power is supplied to the device. 
- HDDs started as massive, room-filling devices with a capacity of about 3.75 megabytes. Today, by comparison, an HDD that fits easily in a desktop computer can have upwards of 18 terabytes worth of storage.
- A solid state drive is non-volatile memory (NVM) computer hardware that stores data without moving parts.
- While SSDs used to be significantly more expensive than HDDs, the cost of NAND flash memory (the storage medium used in SSDs) has decreased over time. This has made SSDs more affordable for a broader range of laptop models.
- 

`EMMC Storage`
- An embedded MultiMediaCard (eMMC) is a small storage device made up of NAND flash memory and a simple storage controller. 
- Commonly used in mobile devices, budget laptops, and other compact electronic devices. 
- Unlike SSDs, which use NAND flash memory chips, eMMC chips are typically soldered directly onto the device's motherboard or integrated into the device's circuitry.
- It's less expensive to manufacture than traditional SSDs
- eMMC storage is faster than traditional HDDs, it is generally slower than SSDs. Sufficient for everyday tasks like web browsing, email, and document editing but may be slower for demanding applications.

`Pixel Per Inches` Ref:**Adobe website and google**
-  PPI refers to the number of individual pixels (tiny dots of light) that are packed into one inch of a screen's surface. It is used to describe the level of detail and sharpness that a display can provide.
- Generally speaking, the higher the PPI, the better the image quality:
    - Lower resolution images contain larger pixels in fewer numbers. This creates a blocky, granular effect hence the term ‘pixelated’.
    - Higher resolution levels benefit from greater numbers of smaller pixels. These create depth, clarity and smoothness.
-  The megapixel measurement is determined by multiplying the horizontal PPI measure by the vertical PPI measure. For instance, a camera that captures images at 1600 x 1200 is considered a 1.92-megapixel camera.
- DPI (dots per inch) is similar to PPI, but is typically used in print media.
- **High-res pixels per inch.**
    - High-res displays can handle more PPI on screen than the commonly held 72 PPI standard. If you’re displaying images or film on a large HD monitor, you can afford a far higher PPI level than your average laptop.
    - A high pixel density can range between 100-140 PPI
    - A very high pixel density starts off at 140 PPI
